*Convert image*

```python
img = cv2.imread('girl.png', 1)

imgGrey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Need to change the gray before do threshold
res,binary = cv2.threshold(imgGrey, 200, 255, cv2.THRESH_BINARY) #src, threshold value, maxVal, thresholdingTechnique

print(res)

plt.subplot(221), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(222), plt.imshow(cv2.cvtColor(imgGrey, cv2.COLOR_BGR2RGB))
plt.subplot(223), plt.imshow(cv2.cvtColor(binary, cv2.COLOR_BGR2RGB))
```
